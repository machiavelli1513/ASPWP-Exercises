```python
import numpy as np
import cupy

def FT2D_numpy(data):
  return  np.fft.ifft2(data)

def FT2D_cupy(data):
  return cupy.fft.ifft2(data)

sizes = [128,256,512,1024,2048]
for size in sizes:
  data_np = np.random.rand(size,size)
  data_cupy = cupy.random.rand(size,size)
  print(f'Numpy size {size}')
  %timeit FT2D_numpy(data_np)
  print(f'Cupy size {size}')
  %timeit FT2D_cupy(data_cupy)
```

```python
# With datatype = np.float32
sizes = [128,256,512,1024,2048]
for size in sizes:
  data_np = np.random.rand(size,size)
  data_np.astype(np.float32)
  data_cupy = cupy.random.rand(size,size)
  data_cupy.astype(cupy.float32)
  print(f'Numpy size {size}')
  %timeit FT2D_numpy(data_np)
  print(f'Cupy size {size}')
  %timeit FT2D_cupy(data_cupy)
```

Answer:
---
As one expects, the larger the size, the longer it takes. At size 256x256, cupy is really starting to outperform numpy.
